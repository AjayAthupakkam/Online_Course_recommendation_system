```python
# 1. USER PROFILE 

# we have to get user intraction for getting similarity
# getting the user_id,course_id,rating,time_spent on course
# renaming the rating and time_spent to avoid the conflict the two features already exits in dataframe 
user_inter = df[['user_id', 'course_id', 'rating', 'time_spent_hours']].copy()
user_inter = user_inter.rename(columns={'rating': 'user_rating',
                                         'time_spent_hours': 'user_time'})

# Creating the Normalize time feature of user spent
user_inter['user_time_norm'] = (
    (user_inter['user_time'] - user_inter['user_time'].min()) /
    (user_inter['user_time'].max() - user_inter['user_time'].min())
)


# creatinng new feature of weight base of rating and time spend
user_inter['weight'] = user_inter['user_rating'] * user_inter['user_time_norm']
user_inter['weight'] = user_inter['weight'].replace(0, 1e-6)

# After getting the weight of each user merge two data Frames on course_id
merged = user_inter.merge(course_dedup.reset_index(), on='course_id', how='left')


# getting all features weights
feature_cols = course_dedup.columns.to_list()
for col in feature_cols:
    merged[col] = merged[col].values * merged['weight'].values

# creating the user profile 
user_profile = (
    merged.groupby('user_id')[feature_cols].sum()
    .div(merged.groupby('user_id')['weight'].sum(), axis=0)
    .fillna(0)
)

print("User Profile Shape:", user_profile.shape)   # (43242, 12)
```

```python
# Getting difficulty level

user_exp = df.groupby('user_id')['previous_courses_taken'].mean().round().astype(int)

def get_difficulty(exp):
    if exp <= 3:    return ['Beginner']
    elif exp <= 10: return ['Beginner', 'Intermediate']
    else:           return ['Intermediate', 'Advanced']

```

```python

# creating course meta data to find out best knowledge based similarity

course_meta = (
    df[['course_id', 'course_name', 'difficulty_level', 'rating', 'course_price']]
    .drop_duplicates('course_id')
    .set_index('course_id')
)

```

```python

# Recommedation function

def recommend(user_id, top_n=5):
    # check if userid exits 
    if user_id not in user_profile.index:
        return f"User {user_id} not found"

    #getting profile of user based on user_id
    profile_vec  = user_profile.loc[user_id].values.reshape(1, -1)
    scores       = cosine_similarity(profile_vec, course_dedup.values)[0]
    score_series = pd.Series(scores, index=course_dedup.index)

    # remove already taken courses
    taken        = df[df['user_id'] == user_id]['course_id'].tolist()
    score_series = score_series.drop(index=taken, errors='ignore')

    # difficulty filter
    exp             = user_exp.get(user_id, 3)
    allowed_courses = course_meta[course_meta['difficulty_level'].isin(get_difficulty(exp))].index
    score_series    = score_series[score_series.index.isin(allowed_courses)]

    # similarity threshold
    score_series = score_series[score_series >= 0.50]

    top            = score_series.sort_values(ascending=False).head(top_n)
    result         = course_meta.loc[top.index].copy()
    result['similarity_score'] = top.values.round(4)

    return result

```

```python

# Evaluation

test_data, train_taken = [], {}

for uid, group in df.groupby('user_id'):
    courses = group['course_id'].tolist()
    if len(courses) >= 2:
        train_taken[uid] = courses[:-1]
        test_data.append({'user_id': uid, 'actual_course': courses[-1]})
    else:
        train_taken[uid] = courses

test_df = pd.DataFrame(test_data)

def recommend_eval(user_id, taken_courses, top_n):
    profile_vec  = user_profile.loc[user_id].values.reshape(1, -1)
    scores       = cosine_similarity(profile_vec, course_dedup.values)[0]
    score_series = pd.Series(scores, index=course_dedup.index)
    score_series = score_series.drop(index=taken_courses, errors='ignore')
    return score_series.sort_values(ascending=False).head(top_n).index.tolist()

sample = test_df.sample(500, random_state=42)
p5, r5, n5, p10, r10, n10 = [], [], [], [], [], []

for _, row in sample.iterrows():
    uid, actual = row['user_id'], row['actual_course']
    if uid not in user_profile.index:
        continue
    taken = train_taken[uid]
    for k, pl, rl, nl in [(5, p5, r5, n5), (10, p10, r10, n10)]:
        recs = recommend_eval(uid, taken, k)
        hit  = 1 if actual in recs else 0
        pl.append(hit / k)
        rl.append(float(hit))
        nl.append((1 / np.log2(recs.index(actual) + 2)) if hit else 0.0)

print("=== EVALUATION RESULTS ===")
print(f"Precision@5  : {np.mean(p5):.4f}")
print(f"Recall@5     : {np.mean(r5):.4f}")
print(f"NDCG@5       : {np.mean(n5):.4f}")
print(f"Precision@10 : {np.mean(p10):.4f}")
print(f"Recall@10    : {np.mean(r10):.4f}")
print(f"NDCG@10      : {np.mean(n10):.4f}")
```

```python
#Test with user id 
print("\n=== USER 7 ===")
print(recommend(7))

print("\n=== USER 500 ===")
print(recommend(500))
```
